```python
# Create a 3x3 array of all zeros
import numpy as np

a = np.zeros((3,3))
print(a)
# Create a 2x2 array of all ones
b = np.ones((2,2))
print(b)
# Create a 3x3 constant array

c = np.full((3,3), 7)
print(c)
# Create a 3x3 array filled with random values
d = np.random.random((3,3))
print(d)
# Create a 3x3 identity matrix
e = np.eye(3)
print(e)
# convert list to array
f = np.array([2, 3, 1, 0])
print(f)
# arange() will create arrays with regularly incrementing values
g = np.arange(20)
print(g)
# note mix of tuple and lists
h = np.array([[0, 1,2.0],[0,0,0],(1+1j,3.,2.)])
print(h)
# create an array of range with float data type
i = np.arange(1, 8, dtype=float)
print(i)
# linspace() will create arrays with a specified number of items which are
# spaced equally between the specified beginning and end values j= np.linspace(start, stop, num)
j = np.linspace(2, 4, 5)
print(j)
```

```python
import numpy as np

arr = np.array([[-1, 2, 0, 4],
[4, -0.5, 6, 0],
[2.6, 0, 7, 8],
[3, -7, 4, 2.0]])

temp = arr[:2, ::2]
print ("Array with first 2 rows and alternate columns(0 and 2):\n", temp)

temp = arr[[0, 1, 2, 3], [3, 2, 1, 0]]
print ("\nElements at indices (0, 3), (1, 2), (2, 1),(3, 0):\n", temp)

cond = arr > 0
temp = arr[cond]
print ("\nElements greater than 0:\n", temp)
```

```python
import numpy as np

# Load dataset (skipping the header row)
data = np.loadtxt("fitness.txt", dtype=str, delimiter="\t", skiprows=1)
# Transpose to split columns easily
data_T = data.T

date = data_T[0]
step_count = np.array(data_T[1], dtype=int)
mood = np.array(data_T[2], dtype=str)
calories_burned = np.array(data_T[3], dtype=int)
hours_of_sleep = np.array(data_T[4], dtype=int)
bool_of_active = np.array(data_T[5], dtype=str)
weight_kg = np.array(data_T[6], dtype=int)

# Multiply step_count and calories_burned element-wise and sum
total_effort = np.dot(step_count, calories_burned)
print("Total Effort:", total_effort)
```
